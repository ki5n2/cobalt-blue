## 데이터프레임과 티블

### select

```r
tib
```

`select(ends_with):` ( )로 끝나는 열

```r
tib %>% select(contains('Y'),everything())
```

### filter

```r
tib
```

`-` X2가 0보다 큰 행

```r
tib %>% filter(X2>0)
```

`-` X2가 0보다 크고 XY2이 0보다 큰 행

```r
tib %>% filter(X2>0, XY2>0)
```

```r
tib %>% filter(X2>0 & XY2>0)
```

`-` 위의 결과에서 XY를 포함하는 열을 추출

```r
tib %>% filter(X2>0 & XY2>0) %>% select(contains("XY"))
```

### mutate

```r
tib
```

`-` 새로운 변수 추가

```r
tib %>% mutate(Z = abs(X2))
```

`-` 여러개의 새로운 변수 추가

### group_by

```r
sex = c("M","M","M","M","F","F","F","F")
value = c(11,12,13,14,21,22,23,24)
```

```r
df = tibble(sex,value)
df
```

```r
df %>% group_by(sex) %>% summarise(mean_value = mean(value))
```

- 그룹별로 묶은 다음 평균을 구함

```r
df %>% group_by(sex) %>% summarise(sd_value = sd(value))
```

- 그룹별로 묶은 뒤 편차를 계산

```r
df %>% group_by(sex) %>% summarise(sqrt_value = sqrt(value))
```

- 원하는 결과가 나오지 않음
- 이는 다음을 수행한 것과 같음 

```r
df %>% mutate(sqrt_value=sqrt(value))
```

`-` `summarise 함수:` 평균과 같이 $n$개의 벡터가 입력으로 올 때 하나의 스칼라 출력을 하는 함수

`-` 따라서 다음은 가능

```r
df %>% group_by(sex) %>% summarise(mean_sqrt_value = mean(sqrt(value)))
```

`-` group_by + mutate

```r
df %>% group_by(sex) %>% mutate(value2=value - mean(value))
```

`-` 위의 코드는 아래와 같음

```r
A1 <- df %>% filter(sex == 'M') %>% mutate(value2 = value - mean(value))
A2 <- df %>% filter(sex == 'F') %>% mutate(value2 = value - mean(value))
```

```r
rbind(A1,A2)
```
